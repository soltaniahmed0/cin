```python
# Install necessary libraries (if not already installed)
#!apt-get install tesseract-ocr -y
#!apt-get install libtesseract-dev -y
#!pip install pytesseract easyocr opencv-python-headless pyocr

# Import necessary libraries
import cv2
import pytesseract
import easyocr
import json
from ultralytics import YOLO
import pyocr
import pyocr.builders
from PIL import Image

# Initialize EasyOCR reader for Arabic language
easyocr_reader = easyocr.Reader(['ar'])  # Specify the languages you want to recognize

# Initialize PyOCR tools
tools = pyocr.get_available_tools()
if len(tools) == 0:
    print("No OCR tool found")
    exit(1)

ocr_tool = tools[0]

# Load a YOLOv8 model (YOLOv8s in this example)
model = YOLO('/content/runs/detect/train2/weights/best (2).pt')

# Function to plot images with bounding boxes (optional)
def plot_image(image, bboxes, texts):
    for (bbox, text) in zip(bboxes, texts):
        (x, y, w, h) = bbox
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow("Image", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Read the input image
image = cv2.imread('C16.jpg')

# Run YOLOv8 inference on the image
results = model(image)

# Extract bounding boxes and class names
bboxes = []
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
        bboxes.append([x1, y1, x2 - x1, y2 - y1])  # Convert to (x, y, w, h) format

# Initialize list to store detected texts from all methods
detected_texts = []

# Loop through the bounding boxes, crop the image, and apply all OCR methods
for bbox in bboxes:
    x, y, w, h = bbox
    cropped_image = image[y:y + h, x:x + w]

    # Convert NumPy array to PIL image
    pil_image = Image.fromarray(cropped_image)

    # Use EasyOCR to read text from the cropped image
    easyocr_results = easyocr_reader.readtext(cropped_image)
    easyocr_text = ' '.join([text for (bbox, text, score) in easyocr_results])

    # Use Tesseract to read text from the cropped image
    tesseract_text = pytesseract.image_to_string(cropped_image, lang='ara').strip()

    # Use Google Tesseract (another model) to read text from the cropped image
    google_tesseract_text = pytesseract.image_to_string(cropped_image, lang='ara', config='--oem 1').strip()

    # Use PyOCR to read text from the cropped image
    pyocr_text = ocr_tool.image_to_string(
        pil_image,
        lang='ara',
        builder=pyocr.builders.TextBuilder()
    ).strip()

    # Store results in a dictionary
    bbox_result = {
        "bbox": bbox,
        "easyocr_text": easyocr_text,
        "tesseract_text": tesseract_text,
        "google_tesseract_text": google_tesseract_text,
        "pyocr_text": pyocr_text
    }
    detected_texts.append(bbox_result)

# Save the results to a JSON file
json_filename = 'text_extraction_results.json'
with open(json_filename, 'w', encoding='utf-8') as f:
    json.dump(detected_texts, f, ensure_ascii=False, indent=4)

print(f"Results saved to {json_filename}")
```

```python
import cv2
import easyocr
import numpy as np  # Import NumPy for array operations
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Initialize EasyOCR reader
reader = easyocr.Reader(['ar'])

# Load the YOLOv8 model
model = YOLO('/content/runs/detect/train/weights/best.pt')

# Function to plot images with bounding boxes
def plot_image(image, bboxes, texts):
    for (bbox, text) in zip(bboxes, texts):
        (x, y, w, h) = bbox
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
    cv2_imshow(image)

# Function to preprocess the image and enhance the text visibility
def preprocess_image(image):
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization to improve contrast
    equalized = cv2.equalizeHist(gray)

    # Apply sharpening to enhance edges
    kernel = np.array([[0,0, 0], [-1, 13, -1], [0, -1, 0]])
    sharpened = cv2.filter2D(equalized, -1, kernel)

    # Apply binary thresholding
    _, binary = cv2.threshold(sharpened, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Invert the image to make text black and background white
    inverted = cv2.bitwise_not(binary)

    return inverted

# Read the input image
image = cv2.imread('test.jpeg')

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not loaded correctly.")
else:
    print("Image loaded successfully.")

# Run YOLOv8 inference on the image
results = model(image)

# Extract bounding boxes and class names
bboxes = []
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
        bboxes.append([x1, y1, x2 - x1, y2 - y1])  # Convert to (x, y, w, h) format

# Initialize list to store detected texts
detected_texts = []

# Loop through the bounding boxes, crop the image, preprocess, and apply EasyOCR
for bbox in bboxes:
    x, y, w, h = bbox

    # Check if the bounding box coordinates are within the image bounds
    if x < 0 or y < 0 or x + w > image.shape[1] or y + h > image.shape[0]:
        print(f"Skipping invalid bounding box: {bbox}")
        continue

    cropped_image = image[y:y + h, x:x + w]

    # Check if the cropped image is valid
    if cropped_image.size == 0:
        print(f"Empty cropped image for bounding box: {bbox}")
        continue

    # Preprocess the cropped image
    processed_image = preprocess_image(cropped_image)

    # Display the processed image for debugging
    cv2_imshow(processed_image)

    # Use EasyOCR to read text from the processed image
    ocr_results = reader.readtext(processed_image)
    detected_text = ' '.join([text for (_, text, _) in ocr_results])
    detected_texts.append(detected_text)

# Plot the image with detected texts
plot_image(image, bboxes, detected_texts)

# Print the results
for bbox, text in zip(bboxes, detected_texts):
    print(f"BBox: {bbox}\nDetected Text: {text}\n")
```
